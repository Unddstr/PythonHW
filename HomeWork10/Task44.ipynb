```python
import pandas as pd
```

```python
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAm': lst})
```

Выведем исходный фрейм:

```python
data.head(20)
```

Переводим фрейм в one hot вид:

```python
df = data.copy(deep=True) #создаем новый фрейм и копируем в него исходный, что бы исходный фрейм остался без изменений
df.loc[data['whoAm'] == 'robot', 'robot'] = "1"
df.loc[data['whoAm'] != 'robot', 'robot'] = "0"
df.loc[data['whoAm'] == 'human', 'human'] = "1"
df.loc[data['whoAm'] != 'human', 'human'] = "0"
df[['human', 'robot']]
```

Для сравнения выведем one hot вид используя get_dummies:

```python
pd.get_dummies(data['whoAm'])
```
